```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../Dataset.csv")

# Task 2: Cuisine Combination
# 1. Identify the most common combinations of cuisines
cuisine_combos = df['Cuisines'].dropna()
combo_counts = cuisine_combos.value_counts()

# Display the top 10 most common cuisine combinations
top_10_combos = combo_counts.head(10)
print("Top 10 Cuisine Combos:")

# Robust rename for different pandas versions
top_10_df = top_10_combos.reset_index()
top_10_df.columns = ['Cuisine Combination', 'Count']
print(top_10_df)

# Plotting the top 10 cuisine combinations
plt.figure(figsize=(12, 6))
sns.barplot(x='Count', y='Cuisine Combination', data=top_10_df, hue='Cuisine Combination', palette='viridis', legend=False)
plt.title('Top 10 Cuisine Combinations')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine Combination')
plt.tight_layout()
plt.show()

# 2. Determine if certain cuisine combinations have higher ratings
top_combos_list = top_10_combos.index.tolist()
top_combo_ratings = df[df['Cuisines'].isin(top_combos_list)]

# Calculate average rating for each of these combinations
avg_rating_per_combo = top_combo_ratings.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
avg_rating_per_combo = avg_rating_per_combo.sort_values(by='Aggregate rating', ascending=False)

# Display the average ratings
print("\nAverage Ratings for Top 10 Cuisine Combinations:")
print(avg_rating_per_combo)

# Plotting average ratings for these combinations
plt.figure(figsize=(12, 6))
sns.barplot(x='Aggregate rating', y='Cuisines', data=avg_rating_per_combo, hue='Cuisines', palette='magma', legend=False)
plt.title('Average Rating for Top 10 Cuisine Combinations')
plt.xlabel('Average Aggregate Rating')
plt.ylabel('Cuisine Combination')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

## 🏁 Conclusion:
- The most frequent cuisine combination is **North Indian**, followed by **North Indian, Chinese**.
- Popular combinations often include a mix of regional and international cuisines (e.g., North Indian & Chinese).
- Some combinations tend to have slightly higher average ratings, suggesting they are either better prepared or more preferred by customers.
