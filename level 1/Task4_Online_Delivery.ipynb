# Level 1 – Task 4: Online Delivery Analysis

## 🔍 Objective:
1. Determine the **percentage of restaurants** that offer online delivery.
2. Compare the **average ratings** of restaurants with and without online delivery.

## 🛠️ Approach:
- Load and clean the dataset.
- Analyze the `Has Online delivery` column.
- Calculate distribution percentages.
- Group by delivery status and calculate mean `Aggregate rating`.
- Visualize with pie and bar charts.

## 📈 Insight:
This helps identify if offering online delivery is associated with better customer satisfaction and how prevalent it is in the market.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("../Dataset.csv")

# 1. Percentage of restaurants offering online delivery
delivery_counts = df['Has Online delivery'].value_counts(normalize=True) * 100
print("Percentage of Restaurants offering Online Delivery:")
print(delivery_counts)

# Visualize Distribution
plt.figure(figsize=(6,6))
plt.pie(delivery_counts, labels=delivery_counts.index, autopct='%1.1f%%', colors=['skyblue', 'orange'])
plt.title("Distribution of Online Delivery")
plt.show()

# 2. Compare average ratings
avg_rating = df.groupby('Has Online delivery')['Aggregate rating'].mean().reset_index()
print("\nAverage Rating Comparison:")
print(avg_rating)

# Visualize Comparison
plt.figure(figsize=(8,5))
sns.barplot(data=avg_rating, x='Has Online delivery', y='Aggregate rating', palette='muted')
plt.title("Average Rating: With vs Without Online Delivery")
plt.ylim(0, 5)
plt.show()
```

## 🏁 Conclusion:
- Approximately **25.7%** of restaurants in the dataset offer online delivery.
- Restaurants that offer online delivery have a significantly higher average rating (**3.25**) compared to those that do not (**2.47**).
- This suggests that online delivery is a valuable feature for both restaurants and customers, often associated with higher-rated establishments.
